## Extra detail

1ï¸âƒ£ Introduction â€“ Kya hai Python Data Classes? ğŸ“˜ğŸ”
Python me agar hume sirf data store karne wali class banani ho (like a model or container), to normally hume manually __init__, __repr__, __eq__ jaise methods likhne padte hain. Lekin Python 3.7 se ek mast feature aaya: @dataclass decorator ğŸ‰

ML, data analysis, or real-world scenarios jahan data object banana hota hai (e.g., Person, ModelMetrics) wahan kaafi useful hota hai.

### ğŸ§ª 2. **Basic Example â€“ Without and With `dataclass`**

#### âš™ï¸ Without `dataclass` (Pure Manual Way):

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Harsh", 22)
print(p1)  # Output: <__main__.Person object at 0x...> âŒ
```

ğŸ‘ Boring output. Na readable, na useful.

#### âœ… With `dataclass`:

```python
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int

p1 = Person("Harsh", 22)
print(p1)  # Output: Person(name='Harsh', age=22) âœ…
```

### ğŸ” Explanation:
- `@dataclass` ne `__init__` & `__repr__` ko automatically generate kar diya.
- Code chhota, readable aur clean ho gaya.

### ğŸ¯ 3. **Extra Examples with Use-Cases**

---

#### ğŸ’¼ **Example 1: Bank Account Dataclass**

```python
from dataclasses import dataclass

@dataclass
class BankAccount:
    holder: str
    account_no: int
    balance: float = 0.0  # default value

acc = BankAccount("Harsh", 101, 5000)
print(acc)
```

â¡ï¸ Automatically prints: `BankAccount(holder='Harsh', account_no=101, balance=5000)`

---

#### ğŸ§® **Example 2: You Can Compare Objects Easily!**

```python
@dataclass
class Product:
    name: str
    price: float

p1 = Product("Book", 200)
p2 = Product("Book", 200)

print(p1 == p2)  # Output: True âœ…
```

ğŸ” Explanation: `dataclass` auto-generates `__eq__()` for comparing content!

---

#### âš¡ **Example 3: Nested Dataclass (ML Style)**

```python
@dataclass
class Model:
    name: str
    accuracy: float

@dataclass
class Experiment:
    id: int
    model: Model

exp1 = Experiment(1, Model("RandomForest", 0.85))
print(exp1)
```

ğŸ’¡ ML me jab models ka data structure bana rahe ho, `dataclass` helps in clean nesting!

---

#### âš™ï¸ **Example 4: Update/Mutate Data**

```python
person = Person("Harsh", 22)
person.age = 23
print(person)
```

ğŸ‘¨â€ğŸ”§ Dataclass objects mutable hote hain unless `frozen=True` diya ho.

#### ğŸ”’ **Example 5: Frozen Dataclass (Immutable Object)**

```python
@dataclass(frozen=True)
class Point:
    x: int
    y: int

p = Point(1, 2)
# p.x = 10  âŒ error: FrozenInstanceError
```

ğŸ“Œ Use-case: Immutable configs, constants, etc.
