## ðŸ§ª **2. Basic Examples with Step-by-Step Explanation**

### ðŸ› ï¸ OOP Constructor Basic Example:

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show_details(self):
        print(f"Brand: {self.brand}, Model: {self.model}")

my_car = Car("Toyota", "Fortuner")
my_car.show_details()
```

ðŸ” **Line-by-Line Explanation:**

- `class Car:` âž¤ Aap ek class bana rahe ho jiska naam Car hai.
- `def __init__...` âž¤ Ye constructor method hai. Jab `Car()` ka object banega, ye function auto-call hoga.
- `self.brand = brand` âž¤ Object ka internal data store ho raha hai.
- `my_car = Car(...)` âž¤ Ek object ban raha hai jisme constructor call ho gaya.
- `show_details()` âž¤ Ek normal function call hai jo object ke values print kar raha hai.

---

### ðŸ§  List Comprehension Basic Example:

```python
numbers = [1, 2, 3, 4, 5]
squared = [n**2 for n in numbers]
print(squared)
```

ðŸ” **Explanation:**

- `n**2 for n in numbers` âž¤ `for` loop laga rahe ho aur `n**2` apply kar rahe ho.
- Ye exactly same hai as:
```python
squared = []
for n in numbers:
    squared.append(n**2)
```
- But pehle wala version zyada clean aur fast hota hai.

### ðŸ§ª Constructor (OOP) Real-World Examples:

#### ðŸŸ¢ Example 1: Multiple Constructor Logic with `*args`

```python
class Animal:
    def __init__(self, *args):
        if len(args) == 1:
            self.name = args[0]
        elif len(args) == 2:
            self.name = args[0]
            self.species = args[1]
        elif len(args) == 3:
            self.name = args[0]
            self.species = args[1]
            self.age = args[2]

dog = Animal("Dog", "Mammal", 5)
print(dog.name, dog.species, dog.age)
```

ðŸ” Yeh flexible constructor hai. Jaise user 1 ya 2 ya 3 values de â€” sab handle ho jaye.

### ðŸ§ª List Comprehension Deep Examples:

#### ðŸŸ¢ Even Number Filter:

```python
nums = [1, 2, 3, 4, 5, 6]
evens = [n for n in nums if n % 2 == 0]
print(evens)
```

#### ðŸŸ¢ Flatten a List of Lists:

```python
lists = [[1, 2], [3, 4], [5, 6]]
flat = [item for sub in lists for item in sub]
print(flat)
```

#### ðŸŸ¢ Square Only Even Numbers:

```python
[n**2 for n in range(1, 11) if n % 2 == 0]
```

#### ðŸŸ¢ Convert Strings to Integers:

```python
strings = ['1', '2', '3']
[int(s) for s in strings]
```

#### ðŸŸ¢ List of Prime Numbers (up to 20):

```python
[n for n in range(2, 20) if all(n % x != 0 for x in range(2, int(n**0.5)+1))]
```
