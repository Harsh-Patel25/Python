### ğŸŒŸ **1. Introduction: Pickling & Unpickling Kya Hota Hai?**  

ğŸ‘‰ Jab hum kisi **Python object ko ek file ya database mein store** karte hain, ya network ke through bhejna hota hai, tab use **byte stream** mein convert karna padta hai.  
Ye process hoti hai **Pickling** yaani **Serialization**.  
Aur jab byte stream ko wapas Python object mein convert karte hain, usse kehte hain **Unpickling** yaani **Deserialization**.

ğŸ“¦ **Pickling** = Python Object â¡ï¸ Byte Stream  
ğŸ“¤ **Unpickling** = Byte Stream â¡ï¸ Python Object

**Use-cases:**
- ML models ko store karna
- Python objects ko share karna
- Session state save karna
- Web applications mein data exchange

### ğŸ”§ **2. Basic Example with `pickle` module**

```python
import pickle

data = {'name': 'ChatGPT', 'type': 'AI Assistant'}

# âœ… Pickling - serialize object and save to file
with open('data.pkl', 'wb') as f:
    pickle.dump(data, f)

# âœ… Unpickling - load object back from file
with open('data.pkl', 'rb') as f:
    loaded_data = pickle.load(f)

print(loaded_data)
```

**Output:**
```
{'name': 'ChatGPT', 'type': 'AI Assistant'}
```

ğŸ§  Yaha `wb` means **write binary** and `rb` means **read binary**.

### ğŸ½ï¸ **3. Real-World Example: ML Dataset Pickling**

```python
import seaborn as sns
import pickle

# ğŸ”¹ Load dataset
df = sns.load_dataset('tips')

# ğŸ”¹ Serialize (Pickle) it to a file
pickle.dump(df, open('file.pkl', 'wb'))

# ğŸ”¹ Deserialize (Unpickle) it from the file
df_loaded = pickle.load(open('file.pkl', 'rb'))

# ğŸ”¹ Print few rows
print(df_loaded.head())
```

ğŸ“‚ Isse tumhare dataset safe ho gaya file mein, bina kisi format loss ke.

### ğŸ“š **4. Dictionary Pickle Example**

```python
dic_example = {'first_name': 'Krish', 'last_name': 'Naik'}

# âœ… Serialize dictionary
pickle.dump(dic_example, open('test.pkl', 'wb'))

# âœ… Deserialize and load
result = pickle.load(open('test.pkl', 'rb'))

print(result)
```

ğŸ§¾ **Output:**
```
{'first_name': 'Krish', 'last_name': 'Naik'}
```

### ğŸš¨ **5. Caution While Using Pickle**

- âŒ Never unpickle data from **untrusted sources** (can contain malicious code)
- âœ… Use only when both sender & receiver are in Python ecosystem
- âœ”ï¸ Use `joblib` for large numerical arrays (scikit-learn models etc.)

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# Train & pickle the model
pickle.dump(model, open('model.pkl', 'wb'))

# Load it back
model_loaded = pickle.load(open('model.pkl', 'rb'))

print(type(model_loaded))
```
