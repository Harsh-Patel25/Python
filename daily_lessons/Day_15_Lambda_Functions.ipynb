### üí° **1. Introduction/Concept Section**

A lambda function in Python is a *mini anonymous function*, jiska **naam nahi hota**, but kaam **full power** ka karta hai ‚ö°

üß† Normally hum functions ko `def` se banate hain, lekin lambda se **ek-line wale function** banate hain. Use-case:  
- Jab temporary function banana ho  
- Jab chhoti si calculation ho  
- Jab kisi function ko argument me function pass karna ho (jaise `map`, `filter`, `sorted` etc.)

üõ† Syntax:
```python
lambda arguments: expression
```

---

### üß™ **2. Basic Example with Step-by-Step Explanation**

```python
f = lambda x, y: x + y
print(f(5, 6))  # ‚ûù 11
```

üîç **Line-by-Line Explanation:**

1. `lambda x, y: x + y` ‚Üí Yahan 2 parameters liye (`x`, `y`) aur return kiya `x + y`
2. `f = lambda...` ‚Üí Is anonymous function ko `f` naam de diya
3. `f(5, 6)` ‚Üí Yani 5 + 6 = **11**

‚ö° Note: `lambda` function **automatically return** karta hai ‚Äî `return` likhne ki zarurat nahi hoti!

### üéØ **3. Extra Examples with Real Use-Cases**

---

#### ‚úÖ Multiply 2 Numbers
```python
multiply = lambda x, y: x * y
print(multiply(3, 4))  # ‚ûù 12
```

#### ‚úÖ Length of String
```python
string_length = lambda s: len(s)
print(string_length("Krish Naik"))  # ‚ûù 10
```

#### ‚úÖ Square all items in a list (using `map`)
```python
numbers = [1, 2, 3, 4, 5, 6]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)  # ‚ûù [1, 4, 9, 16, 25, 36]
```

#### ‚úÖ Filter even numbers (using `filter`)
```python
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # ‚ûù [2, 4, 6]
```

#### ‚úÖ Sort a list of strings by length
```python
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
sorted_fruits = sorted(fruits, key=lambda x: len(x))
print(sorted_fruits)
# ‚ûù ['date', 'apple', 'banana', 'cherry', 'elderberry']
```

### üöÄ **4. Advanced & ML-Style Examples**

---

#### üîç Sort a List of Dictionaries by Age
```python
people = [
    {'name': 'Alice', 'age': 25},
    {'name': 'Bob', 'age': 30},
    {'name': 'Charlie', 'age': 22}
]

sorted_people = sorted(people, key=lambda x: x['age'])
print(sorted_people)
```

üìå Useful in ML when sorting feature dictionaries or dataset rows.

---

#### üîç Max Value in Dictionary using Lambda
```python
data = {'a': 10, 'b': 20, 'c': 5}
print(max(data, key=lambda x: data[x]))  # ‚ûù b
```

‚úÖ Perfect for extracting **max feature** from a dict like feature weights.

---

#### üîç Grouping Words by First Letter using `groupby`
```python
from itertools import groupby
words = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig']
groups = groupby(sorted(words), key=lambda x: x[0])

for key, group in groups:
    print(key, list(group))
```

üß† Helpful in NLP: Grouping words by prefix/starting alphabet.

## üîÅ **More Extra Lambda Function Examples with Explanation**

---

### üîπ **Example 1: Conditional Lambda with if-else (Ternary Operator)**

```python
check_sign = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(check_sign(5))    # ‚ûù Positive
print(check_sign(-2))   # ‚ûù Negative
print(check_sign(0))    # ‚ûù Zero
```

üß† **Explanation:**

- `lambda x:` ‚Üí function 1 argument lega
- `"Positive" if x > 0 else ...` ‚Üí Ternary operator use hua hai
- Ek line me pura `if-elif-else` structure likh diya
- Return type string hoga depending on value

üìå **Use-case:** Data labeling in ML (like sentiment = "positive"/"negative")

---

### üîπ **Example 2: Lambda inside sort() with Nested Conditions**

```python
numbers = [5, 0, -1, -3, 2, 8]

# Sort all positive first, then negative
sorted_numbers = sorted(numbers, key=lambda x: 0 if x >= 0 else 1)
print(sorted_numbers)  # ‚ûù [5, 0, 2, 8, -1, -3]
```

üß† **Explanation:**

- `lambda x: 0 if x >= 0 else 1` ‚Üí agar x positive ya 0 hai to 0, warna 1
- `sorted(..., key=...)` ‚Üí sorting will place 0s first (i.e., positive numbers)

üìå **Use-case:** Custom sorting for display/data cleaning/log grouping.

---

### üîπ **Example 3: Lambda for Multiple Conditions (Chained)**

```python
grade = lambda score: (
    "A+" if score > 90 else
    "A" if score > 80 else
    "B" if score > 70 else
    "C" if score > 60 else
    "Fail"
)

print(grade(95))  # ‚ûù A+
print(grade(75))  # ‚ûù B
print(grade(40))  # ‚ûù Fail
```

üß† **Explanation:**

- `lambda score:` ‚Üí one parameter
- Chain of `if-else` likha for grading logic
- Works like a switch-case structure in one line

üìå **Use-case:** Grading system, evaluation engine in projects or dashboards

---

### üîπ **Example 4: Lambda for Dictionary Sorting by Value Length**

```python
data = {'A': 'apple', 'B': 'banana', 'C': 'kiwi'}
sorted_keys = sorted(data, key=lambda x: len(data[x]))
print(sorted_keys)  # ‚ûù ['C', 'A', 'B']
```

üß† **Explanation:**

- `lambda x: len(data[x])` ‚Üí length of the value
- `sorted(data, key=...)` ‚Üí sort keys based on their value's length

üìå **Use-case:** NLP or textual data where sorting is needed by content size.

---

### üîπ **Example 5: Using Lambda in a Custom Function (ML Style)**

```python
def process_column(data_list, func):
    return [func(i) for i in data_list]

data = [10, 20, 30, 40]
scaled = process_column(data, lambda x: x / 10)
print(scaled)  # ‚ûù [1.0, 2.0, 3.0, 4.0]
```

üß† **Explanation:**

- Function `process_column` takes list + lambda
- Lambda is applied on each item of list
- Real-world ML data scaling example

üìå **Use-case:** Feature engineering / normalization

---

### üîπ **Example 6: Use Lambda for Custom Filtering**

```python
names = ["Krish", "Naik", "Harsh", "Karan", "Ankur", "Amit"]
filtered = list(filter(lambda name: name.startswith("K"), names))
print(filtered)  # ‚ûù ['Krish', 'Karan']
```

üß† **Explanation:**

- `filter(lambda name: name.startswith("K"), names)` ‚Üí keep only names starting with "K"

üìå **Use-case:** NLP / Customer filtering by name patterns
