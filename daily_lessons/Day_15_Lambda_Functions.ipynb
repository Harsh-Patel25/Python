### ğŸ’¡ **1. Introduction/Concept Section**

A lambda function in Python is a *mini anonymous function*, jiska **naam nahi hota**, but kaam **full power** ka karta hai âš¡

ğŸ§  Normally hum functions ko `def` se banate hain, lekin lambda se **ek-line wale function** banate hain. Use-case:  
- Jab temporary function banana ho  
- Jab chhoti si calculation ho  
- Jab kisi function ko argument me function pass karna ho (jaise `map`, `filter`, `sorted` etc.)

ğŸ›  Syntax:
```python
lambda arguments: expression
```

---

### ğŸ§ª **2. Basic Example with Step-by-Step Explanation**

```python
f = lambda x, y: x + y
print(f(5, 6))  # â 11
```

ğŸ” **Line-by-Line Explanation:**

1. `lambda x, y: x + y` â†’ Yahan 2 parameters liye (`x`, `y`) aur return kiya `x + y`
2. `f = lambda...` â†’ Is anonymous function ko `f` naam de diya
3. `f(5, 6)` â†’ Yani 5 + 6 = **11**

âš¡ Note: `lambda` function **automatically return** karta hai â€” `return` likhne ki zarurat nahi hoti!

### ğŸ¯ **3. Extra Examples with Real Use-Cases**

---

#### âœ… Multiply 2 Numbers
```python
multiply = lambda x, y: x * y
print(multiply(3, 4))  # â 12
```

#### âœ… Length of String
```python
string_length = lambda s: len(s)
print(string_length("Krish Naik"))  # â 10
```

#### âœ… Square all items in a list (using `map`)
```python
numbers = [1, 2, 3, 4, 5, 6]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)  # â [1, 4, 9, 16, 25, 36]
```

#### âœ… Filter even numbers (using `filter`)
```python
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # â [2, 4, 6]
```

#### âœ… Sort a list of strings by length
```python
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
sorted_fruits = sorted(fruits, key=lambda x: len(x))
print(sorted_fruits)
# â ['date', 'apple', 'banana', 'cherry', 'elderberry']
```

### ğŸš€ **4. Advanced & ML-Style Examples**

---

#### ğŸ” Sort a List of Dictionaries by Age
```python
people = [
    {'name': 'Alice', 'age': 25},
    {'name': 'Bob', 'age': 30},
    {'name': 'Charlie', 'age': 22}
]

sorted_people = sorted(people, key=lambda x: x['age'])
print(sorted_people)
```

ğŸ“Œ Useful in ML when sorting feature dictionaries or dataset rows.

---

#### ğŸ” Max Value in Dictionary using Lambda
```python
data = {'a': 10, 'b': 20, 'c': 5}
print(max(data, key=lambda x: data[x]))  # â b
```

âœ… Perfect for extracting **max feature** from a dict like feature weights.

---

#### ğŸ” Grouping Words by First Letter using `groupby`
```python
from itertools import groupby
words = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig']
groups = groupby(sorted(words), key=lambda x: x[0])

for key, group in groups:
    print(key, list(group))
```

ğŸ§  Helpful in NLP: Grouping words by prefix/starting alphabet.

### ğŸ”¹ **Example 1: Conditional Lambda with if-else (Ternary Operator)**

```python
check_sign = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(check_sign(5))    # â Positive
print(check_sign(-2))   # â Negative
print(check_sign(0))    # â Zero
```

ğŸ§  **Explanation:**

- `lambda x:` â†’ function 1 argument lega
- `"Positive" if x > 0 else ...` â†’ Ternary operator use hua hai
- Ek line me pura `if-elif-else` structure likh diya
- Return type string hoga depending on value

ğŸ“Œ **Use-case:** Data labeling in ML (like sentiment = "positive"/"negative")

---

### ğŸ”¹ **Example 2: Lambda inside sort() with Nested Conditions**

```python
numbers = [5, 0, -1, -3, 2, 8]

# Sort all positive first, then negative
sorted_numbers = sorted(numbers, key=lambda x: 0 if x >= 0 else 1)
print(sorted_numbers)  # â [5, 0, 2, 8, -1, -3]
```

ğŸ§  **Explanation:**

- `lambda x: 0 if x >= 0 else 1` â†’ agar x positive ya 0 hai to 0, warna 1
- `sorted(..., key=...)` â†’ sorting will place 0s first (i.e., positive numbers)

ğŸ“Œ **Use-case:** Custom sorting for display/data cleaning/log grouping.

---

### ğŸ”¹ **Example 3: Lambda for Multiple Conditions (Chained)**

```python
grade = lambda score: (
    "A+" if score > 90 else
    "A" if score > 80 else
    "B" if score > 70 else
    "C" if score > 60 else
    "Fail"
)

print(grade(95))  # â A+
print(grade(75))  # â B
print(grade(40))  # â Fail
```

ğŸ§  **Explanation:**

- `lambda score:` â†’ one parameter
- Chain of `if-else` likha for grading logic
- Works like a switch-case structure in one line

ğŸ“Œ **Use-case:** Grading system, evaluation engine in projects or dashboards

---

### ğŸ”¹ **Example 4: Lambda for Dictionary Sorting by Value Length**

```python
data = {'A': 'apple', 'B': 'banana', 'C': 'kiwi'}
sorted_keys = sorted(data, key=lambda x: len(data[x]))
print(sorted_keys)  # â ['C', 'A', 'B']
```

ğŸ§  **Explanation:**

- `lambda x: len(data[x])` â†’ length of the value
- `sorted(data, key=...)` â†’ sort keys based on their value's length

ğŸ“Œ **Use-case:** NLP or textual data where sorting is needed by content size.

### ğŸ”¹ **Example 5: Using Lambda in a Custom Function (ML Style)**

```python
def process_column(data_list, func):
    return [func(i) for i in data_list]

data = [10, 20, 30, 40]
scaled = process_column(data, lambda x: x / 10)
print(scaled)  # â [1.0, 2.0, 3.0, 4.0]
```

ğŸ§  **Explanation:**

- Function `process_column` takes list + lambda
- Lambda is applied on each item of list
- Real-world ML data scaling example

ğŸ“Œ **Use-case:** Feature engineering / normalization

### ğŸ”¹ **Example 6: Use Lambda for Custom Filtering**

```python
names = ["Krish", "Naik", "Harsh", "Karan", "Ankur", "Amit"]
filtered = list(filter(lambda name: name.startswith("K"), names))
print(filtered)  # â ['Krish', 'Karan']
```

ğŸ§  **Explanation:**

- `filter(lambda name: name.startswith("K"), names)` â†’ keep only names starting with "K"

ğŸ“Œ **Use-case:** NLP / Customer filtering by name patterns
